```python
x=df_encoded.drop('Fault',axis=1)
```

```python
y=df_encoded['Fault']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y, test_size=.20, random_state=50)
```

```python
x_train.head().T
```

```python
from sklearn import ensemble
rf_clf=ensemble.RandomForestClassifier(n_estimators=180,max_features=21,oob_score=True)
rf_clf.fit(x_train,y_train)
```

```python
y_pred=rf_clf.predict(x_test)
y_pred[100:150]
```

```python
from sklearn import metrics
print('Random Forest Accuracy=' , metrics.accuracy_score(y_pred,y_test))
```

```python
from sklearn.ensemble import GradientBoostingClassifier
gbmodel=GradientBoostingClassifier(max_depth=7,n_estimators=150)############increases n_estimators increases accuracy
gbmodel.fit(x_train, y_train)
y_pred_gb=gbmodel.predict(x_test)
```

```python
print('Grading Boosting Accuracy=' ,metrics.accuracy_score(y_test,y_pred_gb))
```
